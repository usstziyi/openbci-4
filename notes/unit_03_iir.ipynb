**IIR** 的全称是 **Infinite Impulse Response**，中文翻译为**无限脉冲响应**。

简单来说，IIR 是一种数字滤波器的类型，它的核心特征是：

1.  **有反馈**：滤波器的输出不仅依赖于当前的输入和过去的输入，还依赖于**过去的输出**（即反馈回路）。
2.  **脉冲响应无限长**：因为存在反馈，当输入一个单位脉冲信号后，输出理论上会无限延续下去（虽然幅度会逐渐衰减到接近零，但理论上永远不会完全归零）。
3.  **效率高**：相比另一种常见的 FIR（有限脉冲响应）滤波器，IIR 可以用更低的阶数（更少的计算量）实现同样陡峭的滤波特性。

**在你刚才讨论的陷波滤波器场景中：**

- 你使用的 `method='iir'` 就是告诉 MNE 用 **IIR 滤波器** 来设计陷波。
- 这种设计方法直接利用了模拟滤波器的成熟理论（如巴特沃斯、切比雪夫），通过双线性变换映射到数字域，非常适合需要**高计算效率**和**窄带精确陷波**的场景（比如精准滤除 50Hz 工频干扰）。

**对比 FIR：**
- **IIR**：计算快、阶数低、有相位失真（非线性相位）。
- **FIR**：计算慢、阶数高、可实现线性相位（无相位失真）。

在脑电信号处理中，IIR 常用于**实时处理**或**对相位不敏感**的预处理步骤（如陷波），而 FIR 常用于需要**保持波形精确形状**的分析步骤。
