**IRN** 是衡量芯片“底噪”的指标，你可以把它理解为芯片的“本底噪声”。

这直接决定了你的设备**“能看清多小的信号”**。为了让你彻底明白，我们可以用一个**“听诊器”**的例子来类比。

### 什么是输入参考噪声？

想象一下，你是一名医生，用听诊器听病人的心跳：

- **理想情况**：听诊器是绝对安静的，你听到的所有声音都是病人的心跳。
- **现实情况**：听诊器本身因为材质摩擦、环境杂音，会发出细微的**“沙沙”**声。
- **后果**：如果病人的心跳很微弱，而听诊器的“沙沙”声很大，你就听不清心跳了。

**IRN 就是这个“沙沙”声的大小。**

- **定义**：它是指 ADC 芯片内部电路自己产生的噪声。
- **“输入参考”的意思**：为了方便计算，工程师把芯片内部产生的所有噪声，折算成仿佛是从**输入端**进来的噪声电压。
- **单位**：$\mu V_{pp}$。这表示噪声电压的**峰峰值**，也就是噪声波动的最大幅度。

### 为什么强调“70 Hz 带宽”？

噪声是无处不在的，但不同频率的噪声影响不同。

- **带宽**：就像是一个滤网，只允许特定频率范围内的信号通过。
- **70 Hz**：脑电图信号主要集中在低频段（0.5 Hz - 70 Hz）。
- **意义**：这句话的意思是，“在我们最关心的脑电频率范围内（0-70Hz），这个芯片产生的内部噪音非常小”。

### 为什么这个指标很重要？

图片里最关键的一句话是：**“远小于典型 EEG 信号幅值 (10-100 $\mu V$)”**。

这就像是在比**“嗓门大小”**：

- **信号（EEG）**：病人的嗓门。典型脑电信号是 **10 到 100 $\mu V$**。
- **噪声（IRN）**：环境的嘈杂声。这个芯片的噪声只有 **$\le$ 1.0 $\mu V_{pp}$**。

**对比结果**：
因为 **1.0 $\mu V$**（噪声）远远小于 **10 $\mu V$**（最小信号），所以芯片的“沙沙”声完全不会掩盖住脑电波的“说话声”。
